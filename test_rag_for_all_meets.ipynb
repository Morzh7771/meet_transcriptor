```python
def result_(quary, result):
    print(f"Query: {query}\n")
    print(f"Applied filters: {json.dumps(result['applied_filters'], indent=2)}\n")
    print(f"Total results: {result['total_results']}\n")

    if result['results']:
        print("First result:")
        first = result['results'][0]['metadata']
        print(f"  Meeting ID: {first['meeting_id']}")
        print(f"  Client ID: {first['client_id']}")
        print(f"  Consultant ID: {first['consultant_id']}")
        print(f"  Title: {first['title'][:80]}...")
        print(f"  Date: {first['date_start']}")

    print(result['answer'])
```
