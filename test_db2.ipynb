```python
async def populate_database():
    
    
    # Drop and create tables
    # await db.drop_all_tables()
    await db.create_tables()
    
    # 1. Create Client (John Anderson)
    client_data = ClientCreate(
        citizenship="U.S. citizen",
        marital_status="married",
        id_number="123456789",
        id_type="Driver's License",
        country_of_issuance="United States",
        id_issuance_date=datetime(2020, 3, 20),
        id_expiration_date=datetime(2028, 3, 15)
    )
    
    client = await db.create_client(client_data)
    print(f"Created client: {client.id}")
    
    # 2. Create Beneficiaries
    # Primary beneficiary - Sarah Anderson (spouse)
    beneficiary_sarah_data = BeneficiaryCreate(
        client_id=client.id,
        beneficiary_type="primary",
        relation="spouse",
        share_percentage=60.0
    )
    
    beneficiary_sarah = await db.create_beneficiary(beneficiary_sarah_data)
    print(f"Created beneficiary Sarah: {beneficiary_sarah.id}")
    
    # Trust beneficiary
    beneficiary_trust_data = BeneficiaryCreate(
        client_id=client.id,
        beneficiary_type="primary",
        relation="trust",
        share_percentage=40.0
    )
    
    beneficiary_trust = await db.create_beneficiary(beneficiary_trust_data)
    print(f"Created beneficiary Trust: {beneficiary_trust.id}")
    
    # Contingent beneficiaries
    beneficiary_emily_data = BeneficiaryCreate(
        client_id=client.id,
        beneficiary_type="contingent",
        relation="daughter",
        share_percentage=50.0
    )
    
    beneficiary_emily = await db.create_beneficiary(beneficiary_emily_data)
    print(f"Created beneficiary Emily: {beneficiary_emily.id}")
    
    beneficiary_michael_data = BeneficiaryCreate(
        client_id=client.id,
        beneficiary_type="contingent",
        relation="son",
        share_percentage=50.0
    )
    
    beneficiary_michael = await db.create_beneficiary(beneficiary_michael_data)
    print(f"Created beneficiary Michael: {beneficiary_michael.id}")
    
    # 3. Create Persons
    # John Anderson (client person)
    person_john_data = personCreate(
        client_id=client.id,
        beneficiary_id=beneficiary_sarah.id,  # Using Sarah as primary beneficiary
        first_name="John",
        middle_name="Michael",
        last_name="Anderson",
        date_of_birth=datetime(1976, 3, 15),
        sex="Male",
        ssn_or_tin="123-45-6789",
        email="john.anderson@email.com",
        phone_number="+1-212-555-0123",
        phone_alt="+1-631-555-0456"
    )
    
    person_john = await db.create_person(person_john_data)
    print(f"Created person John: {person_john.id}")
    
    # Sarah Anderson (spouse beneficiary)
    person_sarah_data = personCreate(
        client_id=client.id,
        beneficiary_id=beneficiary_sarah.id,
        first_name="Sarah",
        middle_name="Elizabeth",
        last_name="Anderson",
        date_of_birth=datetime(1978, 11, 22),
        sex="Female",
        ssn_or_tin="987-65-4321",
        email="sarah.anderson@email.com",
        phone_number="+1-212-555-0124",
        phone_alt="+1-631-555-0457"
    )
    
    person_sarah = await db.create_person(person_sarah_data)
    print(f"Created person Sarah: {person_sarah.id}")
    
    # Emily Anderson (daughter - contingent beneficiary)
    person_emily_data = personCreate(
        client_id=client.id,
        beneficiary_id=beneficiary_emily.id,
        first_name="Emily",
        middle_name="",
        last_name="Anderson",
        date_of_birth=datetime(2008, 8, 12),
        sex="Female",
        ssn_or_tin="",  # Minor, no SSN recorded
        email="",
        phone_number="",
        phone_alt=""
    )
    
    person_emily = await db.create_person(person_emily_data)
    print(f"Created person Emily: {person_emily.id}")
    
    # Michael Anderson Jr. (son - contingent beneficiary)
    person_michael_jr_data = personCreate(
        client_id=client.id,
        beneficiary_id=beneficiary_michael.id,
        first_name="Michael",
        middle_name="",
        last_name="Anderson Jr.",
        date_of_birth=datetime(2011, 5, 3),
        sex="Male",
        ssn_or_tin="",  # Minor, no SSN recorded
        email="",
        phone_number="",
        phone_alt=""
    )
    
    person_michael_jr = await db.create_person(person_michael_jr_data)
    print(f"Created person Michael Jr: {person_michael_jr.id}")
    
    # 4. Create Person Addresses
    # John's address
    address_john_data = PersonAddressCreate(
        person_id=person_john.id,
        address_type="primary",
        street="785 Park Avenue, Apt 15B",
        city="New York",
        state="NY",
        country="United States",
        zip_code="10021"
    )
    
    address_john = await db.create_person_address(address_john_data)
    print(f"Created address for John: {address_john.id}")
    
    # Sarah's address (same as John's)
    address_sarah_data = PersonAddressCreate(
        person_id=person_sarah.id,
        address_type="primary",
        street="785 Park Avenue, Apt 15B",
        city="New York",
        state="NY",
        country="United States",
        zip_code="10021"
    )
    
    address_sarah = await db.create_person_address(address_sarah_data)
    print(f"Created address for Sarah: {address_sarah.id}")
    
    # Trust address
    address_trust_data = PersonAddressCreate(
        person_id=person_john.id,  # Using John's person_id for trust administrative address
        address_type="trust",
        street="450 Lexington Avenue, Suite 2800",
        city="New York",
        state="NY",
        country="United States",
        zip_code="10017"
    )
    
    address_trust = await db.create_person_address(address_trust_data)
    print(f"Created trust address: {address_trust.id}")
    
    # 5. Create Client Employment
    employment_data = ClientEmploymentCreate(
        client_id=client.id,
        company_name="Goldman Sachs Investment Bank",
        job_title="Senior Vice President, Investment Division",
        job_description="smth",
        hire_date=datetime(2005, 6, 15),
        pay_frequency="Bi-weekly",
        year_funds=550000.0,
        add_funds=200000.0
    )
    
    employment = await db.create_client_employment(employment_data)
    print(f"Created employment: {employment.id}")
    
    # 6. Create Client Education (adding hypothetical education background)
    # MBA from Wharton
    education_mba_data = ClientEducationCreate(
        client_id=client.id,
        started_on=datetime(1998, 9, 1),
        ended_on=datetime(2000, 5, 15),
        field_of_study="Finance and Investment Management",
        degree="Master of Business Administration (MBA)",
        university_name="University of Pennsylvania - Wharton School"
    )
    
    education_mba = await db.create_client_education(education_mba_data)
    print(f"Created MBA education: {education_mba.id}")
    
    # Bachelor's degree
    education_bachelor_data = ClientEducationCreate(
        client_id=client.id,
        started_on=datetime(1994, 9, 1),
        ended_on=datetime(1998, 5, 15),
        field_of_study="Economics",
        degree="Bachelor of Arts (B.A.)",
        university_name="Columbia University"
    )
    
    education_bachelor = await db.create_client_education(education_bachelor_data)
    print(f"Created Bachelor's education: {education_bachelor.id}")
    
    # 7. Create Plans
    # 401(k) Plan
    plan_401k_data = PlanCreate(
        client_id=client.id,
        plan_type="Employer 401(k)",
        provider="Fidelity Investments",
        plan_code="GS-401K-2024",
        plan_name="Goldman Sachs Retirement Plan",
        employer_tax_id="13-5661851",
        roth_first_year=7000.0  # From contribution_settings
    )
    
    plan_401k = await db.create_plan(plan_401k_data)
    print(f"Created 401k plan: {plan_401k.id}")
    
    # Traditional IRA Plan
    plan_ira_data = PlanCreate(
        client_id=client.id,
        plan_type="Traditional IRA",
        provider="Charles Schwab",
        plan_code="IRA-7829456",
        plan_name="Individual Retirement Account",
        employer_tax_id="",  # Not applicable for IRA
        roth_first_year=0.0  # Traditional IRA
    )
    
    plan_ira = await db.create_plan(plan_ira_data)
    print(f"Created plan_ira plan: {plan_ira.id}")
    
    
await populate_database()
```
